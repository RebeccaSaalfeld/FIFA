```python
fifa_data = fifa_data.drop(['Team & Contract','Position', 'Crossing', 'Finishing', 'Heading Accuracy', 'Short Passing', 'Volleys', 'Dribbling', 'Curve', 'FK Accuracy', 'Long Passing', 'Ball Control', 'Acceleration', 'Sprint Speed', 'Agility', 'Reactions', 'Balance', 'Shot Power', 'Jumping', 'Stamina', 'Long Shots', 'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure', 'Marking', 'Standing Tackle', 'Sliding Tackle', 'GK Diving', 'GK Handling', 'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes'], axis=1)
display(fifa_data.head())
```

```python
print(fifa_data['SM'][0])
```

```python
#Cancel the ★ from the lines SM, W/F and IR
fifa_data['SM']=fifa_data['SM'].str.replace('★','') 
fifa_data['W/F']=fifa_data['W/F'].str.replace('★','')
fifa_data['IR']=fifa_data['IR'].str.replace('★','')
```

```python
# Check if the stars are gone:
display(fifa_data[['SM', 'W/F', 'IR']].head())
```

```python
# Replace some of the titles with more understandable titles:
fifa_data = fifa_data.rename(columns={'BP' : 'Best Position',
                                'A/W' : 'Attacking Workrate', 
                                'D/W':'Defensive Workrate', 
                                'IR' : 'International Reputation', 
                                'W/F' : 'Weak Foot', 
                                'SM' : 'Skill Move',
                                'Loan Date End' : 'On Loan'})
```

```python
#Some of the players perform better depending on the opponent team. We can remove these values
#To get rid of the +1 etc. define a function:
def split(x):
    x=x.str.split('+', n=1, expand=True)[0]
    return x
```

```python
# and run it for the affected columns with a for loop:
plus_values=fifa_data.iloc[:,-28:-1]

col_names=[]
for col in plus_values.columns:
    col_names.append(col)
    
print(col_names)
```

```python
cols = fifa_data[['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB', 'GK']]

for i in cols:
     fifa_data[i] = split(cols[i])
        
```

```python
#Convert lbs in kg:
numb2 = []
for i in fifa_data['Weight']:
    i = i.replace('lbs','')
    i = float(i)
    #print(i)
    i = i*0.453592
    print(i)
    numb2.append(round(i,2))
fifa_data['Weight']=numb2
```

```python
# Deal with the columns Value,Wage and Release Clause. Remove the € sign and transform K and M to 1000 and 100,000 respectively
money_col= ['Value', 'Wage', 'Release Clause']
numb3 = []

for col in money_col:
    for i in fifa_data[col]:
        i = i.replace('€','')
        
        if i[-1] == 'K':
            i = i.replace('K','')
            i = int(i) * 1000
            numb3.append(i)
        
        elif i[-1] == 'M':
            i = i.replace('M','')
            i = float(i) * 1000000
            i = int(i)
            numb3.append(i)
            
        else:
            numb3.append(0)
    fifa_data[col] = numb3
    numb3 = []
    
```

```python
#Convert Feet and Inch into meter:
fifa_data['Height in Meter'] = fifa_data['Height'].str.split("'", n = 1, expand = True)[0]
fifa_data['Height 2'] = fifa_data['Height'].str.split("'", n = 0, expand = True)[0]



numb4= []
for i in fifa_data['Height in Meter']:
    i = float(i) * 30.48
    i = int(i)
    numb4.append(i)
fifa_data['Height in Meter'] = numb4


numb5= []
for i in fifa_data['Height 2']:
    i = float(i) * 2.54
    i = int(i)
    numb5.append(i)
fifa_data['Height 2'] = numb5

fifa_data['Height'] = fifa_data['Height in Meter']+fifa_data['Height 2']
```
