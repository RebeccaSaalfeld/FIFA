```python
import numpy as np
import pandas as pd
```

```python
fifa_data = pd.read_csv('fifa21_train.csv')
pd.set_option('display.max_columns', 101)
display(fifa_data.head())
```

```python
# Have a look at the datatypes of the columns:
#pd.set_option('display.max_rows', 101)
#print(fifa_data.dtypes)
```

```python
fifa_data = fifa_data.drop(['Team & Contract','Position', 'Crossing', 'Finishing', 'Heading Accuracy', 'Short Passing', 'Volleys', 'Dribbling', 'Curve', 'FK Accuracy', 'Long Passing', 'Ball Control', 'Acceleration', 'Sprint Speed', 'Agility', 'Reactions', 'Balance', 'Shot Power', 'Jumping', 'Stamina', 'Long Shots', 'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure', 'Marking', 'Standing Tackle', 'Sliding Tackle', 'GK Diving', 'GK Handling', 'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes'], axis=1)
display(fifa_data.head())
```

```python
#To get rid of the stars in the columns SM, W/F and IR print out the star and replace it:
```

```python
print(fifa_data['SM'][0])
```

```python
#Cancel the ★ from the lines SM, W/F and IR
fifa_data['SM']=fifa_data['SM'].str.replace('★','') 
fifa_data['W/F']=fifa_data['W/F'].str.replace('★','')
fifa_data['IR']=fifa_data['IR'].str.replace('★','')
```

```python
# Check if the stars are gone:
display(fifa_data[['SM', 'W/F', 'IR']].head())
```

```python
# Replace some of the titles with more understandable titles:
fifa_data = fifa_data.rename(columns={'BP' : 'Best Position',
                                'A/W' : 'Attacking Workrate', 
                                'D/W':'Defensive Workrate', 
                                'IR' : 'International Reputation', 
                                'W/F' : 'Weak Foot', 
                                'SM' : 'Skill Move',
                                'Loan Date End' : 'On Loan'})
```

```python
#Some of the players perform better depending on the opponent team. We can remove these values
#To get rid of the +1 etc. define a function:
def split(x):
    x=x.str.split('+', n=1, expand=True)[0]
    return x
```

# Dealing with NaN values

```python
#Have a look at the percentage of null values in a column to decide what to do about NaN values:

nulls_fifa_data = pd.DataFrame(round(fifa_data.isna().sum()/len(fifa_data),4)*100) #create a new dataframe that contains the percentages

nulls_fifa_data = nulls_fifa_data.reset_index()

nulls_fifa_data.columns = ['header_name', 'percent_null']
display(nulls_fifa_data[nulls_fifa_data['percent_null']!= 0])
```

```python
print(fifa_data['Loan Date End'].value_counts())
```

```python
fifa_data['Loan Date End']=fifa_data['Loan Date End'].fillna('Jan 01, 1900')
#fifa_data.isna().sum()
```

```python
##print(fifa_data[fifa_data['Position'].isna()].head())
#print(fifa_data['Position'].isna().value_counts())
##print(fifa_data['Position'].unique())
#print(fifa_data['Position'].value_counts())
#
```

```python
#Drop the lines where positions are NaN. There are so many different positions and the lines with NaN are just about 3 %
#display(fifa_data.shape)
fifa_data = fifa_data [fifa_data['Position'].isna()==False]
display(fifa_data.shape)
```

```python
#Have a look at the percentage of null values in a column to decide what to do about NaN values:

nulls_fifa_data = pd.DataFrame(round(fifa_data.isna().sum()/len(fifa_data),4)*100) #create a new dataframe that contains the percentages

nulls_fifa_data = nulls_fifa_data.reset_index()

nulls_fifa_data.columns = ['header_name', 'percent_null']
display(nulls_fifa_data[nulls_fifa_data['percent_null']!= 0])
```

```python
display(fifa_data.shape)
fifa_data = fifa_data [fifa_data['Joined'].isna()==False]
display(fifa_data.shape)
```

```python
display(fifa_data.shape)
fifa_data = fifa_data [fifa_data['Attacking Workrate'].isna()==False]
display(fifa_data.shape)
```
